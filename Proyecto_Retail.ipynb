```python
# Importar las bibliotecas necesarias
import pandas as pd
from google.colab import files  # Biblioteca para cargar archivos en Google Colab
from IPython.display import display  # Para visualizar las tablas formateadas

# Cargar el archivo CSV al entorno de Google Colab
print("Por favor, sube el archivo CSV que deseas analizar (retail_sales_dataset.csv)")
uploaded = files.upload()  # Cuadro de diálogo para cargar el archivo desde el sistema local

# Obtener el nombre del archivo cargado dinámicamente
nombre_archivo = next(iter(uploaded))  # Obtener el primer archivo subido, por si hay varios

# Cargar el archivo CSV en un DataFrame de pandas usando el nombre dinámico
df = pd.read_csv(nombre_archivo)

# Revisar las primeras filas del DataFrame para entender la estructura de los datos
print("Primeras 10 filas del DataFrame para revisar la estructura inicial de los datos:")
display(df.head(10))

# Información general del DataFrame para ver tipos de datos y posibles problemas
print("\nInformación general del DataFrame (tipos de datos y valores nulos):")
df.info()

# Mostrar un resumen de los valores únicos en cada columna para verificar tipos de datos y consistencia
print("\nResumen de los valores únicos en cada columna para identificar problemas:")
for column in df.columns:
    print(f"Columna: {column} - Valores únicos:\n{df[column].unique()[:5]}")  # Mostrar solo los primeros 5 valores únicos

# Asegurarnos de que las columnas no tengan espacios adicionales o inconsistencias
df.columns = df.columns.str.strip().str.lower()

# Renombrar columnas para simplificar su uso en el código
df.rename(columns={
    'transaction id': 'transaction_id',
    'customer id': 'customer_id',
    'product category': 'product_category',
    'price per unit': 'price_per_unit',
    'total amount': 'total_amount'
}, inplace=True)

# Convertir columnas relevantes a formato numérico para evitar errores
# Aplicar conversión a columnas críticas para asegurar que sean numéricas
df['quantity'] = pd.to_numeric(df['quantity'], errors='coerce')
df['price_per_unit'] = pd.to_numeric(df['price_per_unit'], errors='coerce')
df['total_amount'] = pd.to_numeric(df['total_amount'], errors='coerce')

# Verificar si existen valores no numéricos en las columnas que deberían ser numéricas
print("\nVerificación de valores no numéricos en columnas críticas (valores NaN después de conversión):")
print(df[['quantity', 'price_per_unit', 'total_amount']].isna().sum())

# Eliminar filas con valores nulos en columnas numéricas para evitar errores de formato
df = df.dropna(subset=['quantity', 'price_per_unit', 'total_amount'])

# Verificar valores nulos en las columnas categóricas antes de proceder
print("\nVerificación de valores nulos en columnas categóricas:")
print(df[['product_category', 'gender']].isna().sum())

# Eliminar filas con valores nulos en columnas categóricas para evitar problemas de agrupación
df = df.dropna(subset=['product_category', 'gender'])

# 1. Transformación de Datos

# Crear una nueva columna 'Revenue' que represente el ingreso total por venta (Quantity * Price per Unit)
df['revenue'] = df['quantity'] * df['price_per_unit']
print("\nColumna 'Revenue' creada correctamente. Mostrando las primeras 5 filas:")
styled_revenue = df[['product_category', 'gender', 'quantity', 'price_per_unit', 'revenue']].head(5).style.background_gradient(cmap='coolwarm')
display(styled_revenue)  # Mostrar la tabla

# Normalizar la columna 'Quantity' utilizando Min-Max Normalization
df['normalized_quantity'] = (df['quantity'] - df['quantity'].min()) / (df['quantity'].max() - df['quantity'].min())
print("\nColumna 'Normalized Quantity' creada correctamente. Mostrando las primeras 5 filas:")
styled_normalized_quantity = df[['product_category', 'gender', 'quantity', 'normalized_quantity']].head(5).style.background_gradient(cmap='coolwarm')
display(styled_normalized_quantity)  # Mostrar la tabla

# Crear una columna 'Quantity Category' basada en puntos de corte de la columna 'Quantity'
conditions = [
    (df['quantity'] < 10),  # Cantidad baja
    (df['quantity'] >= 10) & (df['quantity'] < 50),  # Cantidad media
    (df['quantity'] >= 50)  # Cantidad alta
]
categories = ['Low', 'Medium', 'High']
df['quantity_category'] = pd.cut(df['quantity'], bins=[-1, 10, 50, float('inf')], labels=categories)
print("\nColumna 'Quantity Category' creada correctamente. Mostrando la distribución de categorías:")
styled_quantity_category = df['quantity_category'].value_counts().to_frame().style.background_gradient(cmap='Blues')
display(styled_quantity_category)  # Visualización de la distribución con colores

# 2. Agrupación y Agregación

# Agrupar por 'product_category' y 'gender' y calcular estadísticas descriptivas de 'quantity' y 'revenue'
print("\nAgrupación por 'product_category' y 'gender' con estadísticas descriptivas de 'quantity' y 'revenue':")
product_gender_group = df.groupby(['product_category', 'gender'], observed=False).agg({
    'quantity': ['sum', 'mean', 'count'],
    'revenue': ['sum', 'mean', 'max'],
    'normalized_quantity': ['mean', 'std']
})
# Verificar si las columnas son numéricas antes de aplicar formato para evitar errores
product_gender_group = product_gender_group.apply(pd.to_numeric, errors='coerce')
styled_product_gender_group = product_gender_group.style.background_gradient(cmap='coolwarm').format("{:.2f}")
display(styled_product_gender_group)  # Mostrar la tabla

# Agrupar por 'product_category' y 'quantity_category' para ver la relación entre categorías y productos
print("\nAgrupación por 'product_category' y 'quantity_category' para analizar la relación entre ventas y categorías de productos:")
category_group = df.groupby(['product_category', 'quantity_category'], observed=False).agg({
    'quantity': ['sum', 'mean', 'count'],
    'revenue': ['sum', 'mean', 'max']
})
# Verificar si las columnas son numéricas antes de aplicar formato para evitar errores
category_group = category_group.apply(pd.to_numeric, errors='coerce')
styled_category_group = category_group.style.highlight_max(color='lightgreen').highlight_min(color='lightcoral').format("{:.2f}")
display(styled_category_group)  # Mostrar la tabla con colores resaltando máximos y mínimos

# 3. Análisis Personalizado con apply

# Crear una columna con la desviación de cada venta respecto a la media de ventas de su grupo (Product Category y Gender)
df['mean_quantity_by_group'] = df.groupby(['product_category', 'gender'])['quantity'].transform('mean')
# Calcular la desviación de cada venta respecto a la media del grupo
df['quantity_deviation'] = df['quantity'] - df['mean_quantity_by_group']
print("\nColumna 'Quantity Deviation' creada correctamente. Mostrando las primeras 5 filas:")

# Asegurar que las columnas utilizadas son numéricas antes de aplicar formato para evitar errores
styled_deviation = df[['product_category', 'gender', 'quantity', 'mean_quantity_by_group', 'quantity_deviation']].head(10)
# Convertir las columnas necesarias a formato numérico para evitar errores de formato
styled_deviation = styled_deviation.apply(pd.to_numeric, errors='coerce')
styled_deviation = styled_deviation.style.background_gradient(cmap='viridis').format("{:.2f}")
display(styled_deviation)  # Mostrar la tabla


```
